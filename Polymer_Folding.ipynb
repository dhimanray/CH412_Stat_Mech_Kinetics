```python
# @title Polymer Unfolding

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def plot_population(E_misfolded,E_unfolded,g_folded,g_misfolded,g_unfolded):

  k = 1.0

  T = np.arange(0.02,5.0,0.02)

  E_folded = 0.0

  pop_folded = g_folded * np.exp(-E_folded/(k*T))
  pop_misfolded = g_misfolded * np.exp(-E_misfolded/(k*T))
  pop_unfolded = g_unfolded * np.exp(-E_unfolded/(k*T))

  Q = pop_folded + pop_misfolded + pop_unfolded

  p_folded = pop_folded/Q
  p_misfolded = pop_misfolded/Q
  p_unfolded = pop_unfolded/Q

  fig, axes = plt.subplots(1, 2, figsize=(14, 5)) # Create a figure with 2 subplots

  # Plot 1
  axes[0].plot(T,p_folded,label='Folded',linewidth=2)
  axes[0].plot(T,p_misfolded,label='Misfolded',linewidth=2)
  axes[0].plot(T,p_unfolded,label='Unfolded',linewidth=2)
  axes[0].legend(fontsize=12)
  axes[0].set_title('Population of States',fontsize=12,fontweight='bold')
  axes[0].tick_params(axis='x', labelsize=12)
  axes[0].tick_params(axis='y', labelsize=12)
  axes[0].set_xlabel('Temperature (T)',fontsize=14)
  axes[0].set_ylabel('Probability',fontsize=14)

  # Plot 2
  axes[1].plot(T,p_folded,label='Folded',linewidth=2)
  axes[1].plot(T,p_misfolded + p_unfolded,label='Not Folded',linewidth=2)
  axes[1].legend(fontsize=12)
  axes[1].set_title('Probability of Folded and Not Folded',fontsize=12,fontweight='bold')
  axes[1].tick_params(axis='x', labelsize=12)
  axes[1].tick_params(axis='y', labelsize=12)
  axes[1].set_xlabel('Temperature (T)',fontsize=14)
  axes[1].set_ylabel('Probability',fontsize=14)

  #find intersection for Plot 2
  T_m = 0.0
  # Find the index where the difference is minimized, which is close to zero
  idx = np.argmin(np.abs(p_folded - (p_misfolded + p_unfolded)))
  if np.abs(p_folded[idx] - (p_misfolded[idx] + p_unfolded[idx])) < 0.1: # A more robust check for intersection validity
    T_m = T[idx]

  if T_m == 0.0:
    axes[1].text(T[0], 0.5, r'$T_m$ is out of range', va='center') # Adjust text position
  else :
    axes[1].axvline(x=T_m,color='k',ls='--')
    axes[1].text(T_m+0.2,0.9,r'$T_m$ = %0.2f'%T_m,fontsize=14)

  plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
  plt.show()


interact(plot_population,
         E_misfolded=FloatSlider(min=0.1, max=2, step=0.1, value=1.0),
         E_unfolded=FloatSlider(min=2.1, max=4, step=0.1, value=3),
         g_folded=FloatSlider(min=1, max=25, step=1, value=4),
         g_misfolded=FloatSlider(min=1, max=25, step=1, value=11),
         g_unfolded=FloatSlider(min=1, max=25, step=1, value=21))
```
