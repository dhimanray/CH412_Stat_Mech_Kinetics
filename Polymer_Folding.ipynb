```python
# @title Hidden Code
from ipywidgets import interact, FloatSlider
import numpy as np
import matplotlib.pyplot as plt

def plot_population(E_misfolded,E_unfolded,g_folded,g_misfolded,g_unfolded):

  k = 1.0

  T = np.arange(0.02,5.0,0.02)

  E_folded = 0.0

  pop_folded = g_folded * np.exp(-E_folded/(k*T))
  pop_misfolded = g_misfolded * np.exp(-E_misfolded/(k*T))
  pop_unfolded = g_unfolded * np.exp(-E_unfolded/(k*T))

  Q = pop_folded + pop_misfolded + pop_unfolded

  p_folded = pop_folded/Q
  p_misfolded = pop_misfolded/Q
  p_unfolded = pop_unfolded/Q

  plt.figure()
  plt.plot(T,p_folded,label='Folded')
  plt.plot(T,p_misfolded,label='Misfolded')
  plt.plot(T,p_unfolded,label='Unfolded')
  plt.legend()

  plt.figure()
  plt.plot(T,p_folded,label='Folded')
  plt.plot(T,p_misfolded + p_unfolded,label='Not Folded')
  plt.legend()
  #find intersection
  T_m = 0.0
  for t in range(len(T)):
    if np.abs(p_folded[t] - (p_misfolded[t] + p_unfolded[t])) < 1e-2:
      T_m = T[t]
  if T_m == 0.0:
    plt.text(T_m+0.2,0.5,r'$T_m$ is out of range')
  else :
    plt.axvline(x=T_m,color='k',ls='--')
    plt.text(T_m+0.2,0.9,r'$T_m$ = %0.2f'%T_m)




interact(plot_population,
         E_misfolded=FloatSlider(min=0.1, max=2, step=0.1, value=1.0),
         E_unfolded=FloatSlider(min=2.1, max=4, step=0.1, value=3),
         g_folded=FloatSlider(min=1, max=25, step=1, value=4),
         g_misfolded=FloatSlider(min=1, max=25, step=1, value=11),
         g_unfolded=FloatSlider(min=1, max=25, step=1, value=21))
```
