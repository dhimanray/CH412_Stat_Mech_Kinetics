```python
# @title Code
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def harmonic_oscillator(theta):

  # Constants (normalized for visualization)

  kb = 1.0


  # Temperature range (avoiding zero to prevent division by zero)
  T = np.linspace(0.1, 4.0, 400)
  beta = 1 / (kb * T)
  x = theta / T

  # Internal Energy U
  # U = hbar*omega * (0.5 + 1 / (exp(x) - 1))
  internal_energy = theta * kb * (0.5 + 1 / (np.exp(x) - 1))

  # Heat Capacity Cv
  # Cv = kb * x^2 * exp(x) / (exp(x) - 1)^2
  cv = kb * (x**2) * np.exp(x) / (np.exp(x) - 1)**2

  # Plotting
  fig, ax1 = plt.subplots(figsize=(10, 6))

  # Internal Energy Plot
  color1 = 'tab:blue'
  ax1.set_xlabel('Temperature (T)',fontsize=16)
  ax1.set_ylabel('Internal Energy (U)', color=color1,fontsize=16)
  ax1.plot(T, internal_energy, color=color1, lw=2, label='Internal Energy')
  ax1.plot(T, kb*T, color=color1, lw=2, linestyle='--')
  ax1.tick_params(axis='y', labelcolor=color1,labelsize=14)
  ax1.tick_params(axis='x', labelcolor='k',labelsize=14)
  ax1.grid(True, alpha=0.3)

  # Heat Capacity Plot (Secondary Axis)
  ax2 = ax1.twinx()
  color2 = 'tab:red'
  ax2.set_ylabel('Heat Capacity (Cv)', color=color2,fontsize=16)
  ax2.plot(T, cv, color=color2, lw=2, label='Heat Capacity')
  ax2.plot(T, kb*np.ones(len(T)), color=color2, lw=2, linestyle='--')
  ax2.tick_params(axis='y', labelcolor=color2,labelsize=14)

  plt.axvline(x=theta,linestyle='--',color='k',lw=2)
  plt.text(theta+0.2,0.0,r'$\Theta_{vib}$=%0.2f'%theta,fontsize=16)
  plt.text(theta+0.1,0.2,'Classical\nRegime',fontsize=16)
  plt.text(theta-0.6,0.2,'Quantum\nRegime',fontsize=16)


  plt.title('Exact Quantum Result: Solid Line;  Classical Limit: Dashed Line',fontsize=16)
  fig.tight_layout()
  plt.show()

interact(harmonic_oscillator,
         theta=FloatSlider(min=1, max=3, step=0.1, value=1.0))
```
